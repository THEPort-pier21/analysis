```python
import json
from pprint import pprint

with open('data/InsIns_source.json') as data_file:    
    data = json.load(data_file)
```

```python
for l in data:
    print l.get('eventid')
```

```python
event_source= {}
#Key: event id, value: source json object
for source in data:
    if source.get('eventid') in event_source.keys():
        event_source[source.get('eventid')].append(source)
    else:
        event_source[source.get('eventid')] = [source]
    
```
